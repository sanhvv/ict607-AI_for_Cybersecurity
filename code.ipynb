```python
# Selecting Features that are relevant to the DOS attacks.
dos_features = [
    # Network Traffic Volume/Rate
    'dur', 'sbytes', 'dbytes', 'Spkts', 'Dpkts', 'Sload', 'Dload',
    # Transport Layer Behavior
    'sttl', 'dttl', 'tcprtt', 'synack', 'ackdat', 'is_sm_ips_ports',
    'sloss', 'dloss', 'dsport',
    # Application/Session Layer Indicators
    'service', 'state', 'proto', 'ct_flw_http_mthd', 'is_ftp_login', 'ct_ftp_cmd', 'res_bdy_len',
    # Behavioral/Statistical Patterns
    'smeansz', 'dmeansz', 'trans_depth', 'Sjit', 'Djit', 'Sintpkt', 'Dintpkt',
    # Connection-Based
    'ct_srv_dst', 'ct_dst_ltm', 'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm',
    # Labels
    'attack_cat', 'Label'
]
```

```python
# Drop columns not in dos_features
columns_to_drop = [col for col in train_df.columns if col not in dos_features]
train_df = train_df.drop(columns=columns_to_drop)

# Display the first few rows to verify the changes
print(train_df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize separate encoders for each column
proto_encoder = LabelEncoder()
state_encoder = LabelEncoder()
service_encoder = LabelEncoder()
attack_encoder = LabelEncoder()

# Apply label encoding
train_df['proto_encoded'] = proto_encoder.fit_transform(train_df['proto'])
train_df['state_encoded'] = state_encoder.fit_transform(train_df['state'])
train_df['service_encoded'] = service_encoder.fit_transform(train_df['service'])
train_df['attack_type_encoded'] = attack_encoder.fit_transform(train_df['attack_cat'])

# Display sample of encoded results
print(train_df[['proto', 'proto_encoded', 'state', 'state_encoded', 'service', 'service_encoded', 'attack_cat', 'attack_type_encoded']].head())

# Show mapping of original values to encoded values
print("\nProto unique values and their encoding:")
print(dict(zip(proto_encoder.classes_, proto_encoder.transform(proto_encoder.classes_))))

print("\nState unique values and their encoding:")
print(dict(zip(state_encoder.classes_, state_encoder.transform(state_encoder.classes_))))

print("\nService unique values and their encoding:")
print(dict(zip(service_encoder.classes_, service_encoder.transform(service_encoder.classes_))))

print("\nAttack type unique values and their encoding:")
print(dict(zip(attack_encoder.classes_, attack_encoder.transform(attack_encoder.classes_))))
```

```python
# Fix 15,000 samples per class

import pandas as pd
# Define the desired number of samples per class
samples_per_class = 15000

# Group the DataFrame by the 'attack_type_encoded' column
grouped = train_df.groupby('attack_type_encoded')

# Initialize an empty list to store the resampled DataFrames
resampled_dataframes = []

# Iterate through each group (each class)
for group_name, group_data in grouped:
    # Check if the current group has more samples than the desired amount
    if len(group_data) > samples_per_class:
        # If yes, randomly sample the desired number of samples without replacement
        resampled_group = group_data.sample(n=samples_per_class, replace=False, random_state=42) # Use random_state for reproducibility
    else:
        # If no, use all samples in the group
        resampled_group = group_data
    # Append the resampled group to the list
    resampled_dataframes.append(resampled_group)

# Concatenate all the resampled DataFrames back into a single DataFrame
balanced_train_df = pd.concat(resampled_dataframes, ignore_index=True)

# Verify the number of samples per class in the balanced DataFrame
balanced_attack_cat_counts = balanced_train_df['attack_type_encoded'].value_counts()
print("\nCounts of each attack category after balancing:")
balanced_attack_cat_counts
```

```python
# Selected feature columns
feature_cols = [
    'dsport', 'dur', 'sbytes', 'dbytes', 'sttl', 'dttl', 'sloss', 'dloss',
    'Sload', 'Dload', 'Spkts', 'Dpkts', 'smeansz', 'dmeansz', 'trans_depth',
    'res_bdy_len', 'Sjit', 'Djit', 'Sintpkt', 'Dintpkt', 'tcprtt', 'synack',
    'ackdat', 'is_sm_ips_ports', 'ct_flw_http_mthd', 'is_ftp_login',
    'ct_srv_dst', 'ct_dst_ltm', 'ct_src_dport_ltm', 'ct_dst_sport_ltm',
    'ct_dst_src_ltm', 'proto_encoded', 'state_encoded', 'service_encoded'
]

# Define X and binary y (1 = DoS, 0 = not DoS)
X = balanced_train_df[feature_cols]
y = balanced_train_df['attack_type_encoded']
```

```python
# Split the data into training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#Random Forest model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train_scaled, y_train)

# Predict
y_pred = clf.predict(X_test_scaled)

# Evaluation
print("Classification Report (DoS vs Others):")
print(classification_report(y_test, y_pred))


import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Define class labels
labels = sorted(list(set(y_test)))  # or use your custom class names

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
# SVM model
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# classification model
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train_scaled, y_train)

# Predict
y_pred_svm = svm_model.predict(X_test_scaled)

# Evaluation
print("Classification Report SVM (DoS vs Others):")
print(classification_report(y_test, y_pred_svm))

# Compute the confusion matrix
cm_svm = confusion_matrix(y_test, y_pred_svm)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
# Install XGBoost
!pip install xgboost

import xgboost as xgb

# Define and train the XGBoost classifier
xgb_model = xgb.XGBClassifier(objective='multi:softmax',
                              num_class=len(labels),
                              eval_metric='mlogloss',
                              use_label_encoder=False,
                              random_state=42)

xgb_model.fit(X_train_scaled, y_train)

# Predict
y_pred_xgb = xgb_model.predict(X_test_scaled)

# Evaluation
print("Classification Report XGBoost:")
print(classification_report(y_test, y_pred_xgb))

# Compute the confusion matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)

plt.title("Confusion Matrix (XGBoost)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
# Comparison between models
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Get classification report as dictionaries for SVM, Random Forest, and XGBoost
report_svm = classification_report(y_test, y_pred_svm, output_dict=True)
report_clf = classification_report(y_test, y_pred, output_dict=True)
report_xgb = classification_report(y_test, y_pred_xgb, output_dict=True)


# Extract metrics for each class
classes = sorted([str(i) for i in set(y_test)]) # Dynamically get class labels from y_test

svm_precision = [report_svm[c]['precision'] if c in report_svm else 0 for c in classes]
svm_recall    = [report_svm[c]['recall']    if c in report_svm else 0 for c in classes]
svm_f1        = [report_svm[c]['f1-score']  if c in report_svm else 0 for c in classes]

clf_precision = [report_clf[c]['precision'] if c in report_clf else 0 for c in classes]
clf_recall    = [report_clf[c]['recall']    if c in report_clf else 0 for c in classes]
clf_f1        = [report_clf[c]['f1-score']  if c in report_clf else 0 for c in classes]

xgb_precision = [report_xgb[c]['precision'] if c in report_xgb else 0 for c in classes]
xgb_recall    = [report_xgb[c]['recall']    if c in report_xgb else 0 for c in classes]
xgb_f1        = [report_xgb[c]['f1-score']  if c in report_xgb else 0 for c in classes]


# extract accuracy
accuracy_svm = accuracy_score(y_test, y_pred_svm)
accuracy_clf = accuracy_score(y_test, y_pred)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)


# === Plot function for accuracy comparison ===
def plot_accuracy(acc_svm, acc_rf, acc_xgb):
    models = ['SVM', 'Random Forest', 'XGBoost']
    accuracies = [acc_svm, acc_rf, acc_xgb]

    plt.figure(figsize=(7, 5))
    bars = plt.bar(models, accuracies, color=['blue', 'green', 'orange'])

    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, height + 0.01, f'{height:.2f}', ha='center', va='bottom')

    plt.ylim(0, 1.05)
    plt.ylabel('Accuracy')
    plt.title('Overall Accuracy Comparison: SVM vs RF vs XGBoost')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# === Plot function for per-class metrics ===
def plot_comparison(metric_name, svm_scores, rf_scores, xgb_scores):
    x = np.arange(len(classes))
    width = 0.25

    fig, ax = plt.subplots(figsize=(14, 6))
    rects1 = ax.bar(x - width, svm_scores, width, label='SVM', color='blue')
    rects2 = ax.bar(x, rf_scores, width, label='Random Forest', color='green')
    rects3 = ax.bar(x + width, xgb_scores, width, label='XGBoost', color='orange')

    ax.set_ylabel(metric_name)
    ax.set_xlabel('Class Label')
    ax.set_title(f'{metric_name} by Class: SVM vs RF vs XGBoost')
    ax.set_xticks(x)
    ax.set_xticklabels(classes)
    ax.set_ylim(0, 1.1)
    ax.legend()

    # Annotate bars
    for rects in [rects1, rects2, rects3]:
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

# === Plot all ===
plot_accuracy(accuracy_svm, accuracy_clf, accuracy_xgb)
plot_comparison("Precision", svm_precision, clf_precision, xgb_precision)
plot_comparison("Recall", svm_recall, clf_recall, xgb_recall)
plot_comparison("F1-Score", svm_f1, clf_f1, xgb_f1)
```
